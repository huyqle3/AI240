```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 5))
sns.histplot(iris, x='petal_length', kde=True, color='skyblue')
plt.title("Distribution of Petal Length (Iris Dataset)", fontsize=14)
plt.xlabel("Petal Length")
plt.ylabel("Count")
plt.show()

penguins = sns.load_dataset('penguins')

plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='flipper_length_mm', kde=True, color='salmon')
plt.title("Distribution of Flipper Length (Penguins Dataset)", fontsize=14)
plt.xlabel("Flipper Length")
plt.ylabel("Count")
plt.show()
```

# Exercise 6 - Draw a Swarm Plot

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

Create a Swarm plot of Petal length across the species.
 
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors)

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

iris = sns.load_dataset("iris")

species_colors = ['#78C850',  # Setosa
                  '#F08030',  # Versicolor
                  '#6890F0']  # Virginica

plt.figure(figsize=(8, 6))
sns.swarmplot(
    x='species',
    y='petal_length',
    hue='species',
    data=iris,
    palette=species_colors,
    legend=False
)

plt.title("Swarm Plot of Petal Length by Species", fontsize=14)
plt.xlabel("Species")
plt.ylabel("Petal Length")

plt.show()
```

# Exercise 7 - Draw a Bar Plot

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.
Execute below command

sns.barplot(penguins, x="island", y="body_mass_g")


```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8, 6))
sns.barplot(
    data=penguins,
    x="island",
    y="body_mass_g",
    hue="island",
    palette="pastel",
    legend=False
)

plt.title("Average Body Mass of Penguins per Island", fontsize=14)
plt.xlabel("Island")
plt.ylabel("Body Mass")

plt.show()
```

# Exercise 8 - Draw a Heatmap Plot

corrmat = iris.corr()

sns.heatmap(corrmat, annot = True, square = True)


```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

iris = sns.load_dataset("iris")

corrmat = iris.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corrmat, annot=True, square=True, cmap="coolwarm")

plt.title("Correlation Heatmap of Iris Features", fontsize=14)

plt.show()
```
