## Exercise 3 - Draw Barchart




```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("car data.csv")

cars_subset = data.head(5)

plt.bar(cars_subset["Car_Name"], cars_subset["Selling_Price"])

plt.title("Car Selling Price Comparison")
plt.xlabel("Car Model")
plt.ylabel("Selling Price")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Exercise 4 -  Draw Histogram


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("car data.csv")

kms = data["Kms_Driven"]

plt.hist(kms, bins=15, color='skyblue', edgecolor='black', alpha=0.7)

plt.xlabel("Kilometers Driven")
plt.ylabel("Number of Cars")
plt.title("Distribution of Kms Driven")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Exercise 5 -  Draw Piechart



```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("car data.csv")

fuel_counts = data["Fuel_Type"].value_counts()

plt.pie(
    fuel_counts, 
    labels=fuel_counts.index, 
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'skyblue', 'lightgreen']
)

plt.title("Fuel Type Distribution of Cars")

plt.axis('equal')

plt.show()
```
